## 1. Setup and Installation

Because Streamlit apps run as standalone web servers, they cannot execute directly *inside* a Jupyter Notebook cell. Instead, we write the code to a file and run it from the terminal.

### 💡 How to Run the App
To view your application, open your terminal/command prompt, navigate to this folder, and run:
```bash
streamlit run app.py
```
*Note: The app will open automatically in your local web browser.*
