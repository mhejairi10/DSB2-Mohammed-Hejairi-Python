# Part 2: Collections

There are also data types that allow you to *collect* multiple objects together; these are termed collections. There are four common types of collections in Python:

* `list` - `[1, 2, 3]`
* `tuple` - `(1, 2, 3)`
* `dict` - `{'apple': 0.5, 'orange': 0.8, 'pear': 1.3}`
* `set` - `{'this', 'that', 'other'}`

*Note: Strictly speaking, `str` is a collection of individual characters.*

## `List` 

`list` and `tuple` are *linear collections*. Like lines, they consist of elements placed in some order. So, for example, in the `list` `[53, 62, 98]`, the first element is `53`, the second element is `62` and the third is `98`. We can access elements with `[]`, for example:

````python
students = ["Aysha","Hajar","Fajer","Ahmed"]

students
````

````python
my_list = [53, 62, 98 , "Mohammed"]


````

````python
# What is the first element in the list my_list?
my_list[0]

````

Accessing an element with `[]` is called *indexing*. In Python, indices start from 0, so the first element can be accessed with `[0]`, as above, the second with `[1]`, and so on. You can also use *negative* indices to index from the back. So, for example, the last element would be accessed with `[-1]`:

````python
# What is the last element in the list my_list?

my_list[-1]
````

## `Tuple`
A `tuple` is exactly the same as a `list`, with one exception: `tuples` are *immutable*, meaning that you can't change what they contain after you create them.

<font color='blue'>*Tuple is written with Parentheses ()*</font>

````python
my_tuple = (5,8, 62,22,343,987)
````

````python
# You can access tuple elements with indexing, similiar to the list

my_tuple[2]
````

````python
# But you can't modify it, try using .append()

my_tuple.append(3)
````

````python
# tuple does not support item assignment

my_tuple[6]=3
````

````python
# Notice that tuple can contain different data types
one_more_tuple = (32,'Husain',[76,978],54)


````

### Change an internal list that is inside a tuple(Optional)
Note that you can't change what they contain, but the things that they contain can themselves change! For example, imagine you have a `tuple` containing a `list`:

````python
another_tuple = (5, [34, 26], 9)
````

````python
# You can't change what the second element is...

another_tuple[1]=3
````

````python
# ...but you can change what's in it!
# try to append integer 37 to the list that is inside the tuple 
# using .append()
another_tuple[1].append(37)

````

````python
# removing value 26 from a list inside a tuple using .remove()

another_tuple[1].remove(26)

````

## `Dictionary` 

`dict` is used in cases where data has no meaningful order, but each value can be reasonably associated with a key. For example, you could represent the amount of each type of food you have in stock with a `dict`:

```python
{
    'bread': 16,
    'peanuts': 36,
    'fish': 9
}
```

Notice that when you define a `dict`, it consists of pairs of objects separated by colons (`:`). We call the object before the colon the *`key`*, and the object after the *`value`*.

````python
my_dict = {'x': 'apple', 
           'y': 'banana', 
           'z': 'carrot'}
````

Recall that `lists` and `tuples` have order and that you can find their "first" element, their "last" element, etc.. This does not apply to `dicts`, for example:

````python
# Can I get the first element using indexing [0]?

my_dict[0]

````

Instead, a `dict` *associates* one object with another, which is why we call it an *associative collection*.

We can access values in a `dict` by their keys:

````python
my_dict
````

````python
# what is the value of the key x?

my_dict['x']
````

````python
# what is the value of the key y?

my_dict['y']
````

````python
# what is the value of the key z?

my_dict['z']
````

````python
# other way to get the value out of the key by using .get()

my_dict.get('x')
````

However, there is no quick way to access a key by its value. There are technical reasons for this, which are not so important now, but note that keys must be unique, whereas values need not be:

````python
my_dict
````

````python
# add the key 'd' corresponding to the value 'orange' into my_dict

my_dict['d']='orange'

my_dict
````

### keys must be unique

````python
# Only 'alligator' will make it in!

test1 = {'product': 'apple', 'product': 'alligator'}

print(test1)
````

````python
# But this is fine

test2 = {'cheese': 'milk', 'yoghurt': 'milk'}

print(test2)
````

One thing to note is that keys must be *immutable*. In practical terms, that means that you **can't use** `lists`, `dicts` or `sets` for keys:

````python
error = {[36, 87]: 'a'}
````

There's no similar restriction for values:

````python
no_error = {'a': [764, 345]}


````

You **can use** `strings` and `numbers` for keys

````python
no_error2 = {5: [764, 345]}


````

##  `Set`
A `set` is like a `dict` that has no values, only keys. 
- Set is an unordered collection of unique values.


There are some other operations you can perform on `sets` (see all of them [here](https://docs.python.org/3/library/stdtypes.html#set)):

*Note: `sets` are unordered, so you cannot expect that the result of these operations will have any particular ordering.*

In Python sets are written with curly brackets {}

````python
primary_colors = {"red","blue","yellow"}
````

````python
# add an item 'green' using .add()

primary_colors.add("green")
````

````python
# remove an item 'yellow' using .remove()

primary_colors.remove("yellow")
````

Adding something that already exists in a `set` to it will have no effect:
   

````python
primary_colors
````

````python
# adding blue color to the same set does not have an effect 
# since blue already exists in the set

primary_colors.add('blue')
````

### Using `Set` to remove duplicates

#### Tip💡

    
* One common usage of sets are to remove duplicates in a string or a list.
    
* Set is useful for finding the unique values in some object, for example, the unique letters in a `str`:

````python
word = 'abracadabra'

# remove duplicate letters in the variable word by converting to a set

set(word)

````

````python
# assume I want to remove the duplicate values from the follwing list:
mylist=[4, 5, 6, 7, 8, 9, 10,5,6,9]

# get the unique items in a list by casting/converting the list to a set
# using set()

mylist = set(mylist)
````

````python
# you can see that mylist is converted to a set now
# and does not have duplicates at all
mylist
````

````python
# Convert back to a list

mylist = list(mylist)
mylist
````

````python
# You can test if a value exists in a set or a list using in
x = 9

````

````python

x in mylist
````

### Intersection of Sets

````python
even_numbers = {0, 2, 4, 6, 8, 10}
more_than_three = {4, 5, 6, 7, 8, 9, 10}
````

````python
# show intersection of even_numbers and more_than_three using &

print(even_numbers & more_than_three)
````

### Union of Sets

````python
# either even_numbers or more_than_three using |

print(even_numbers | more_than_three)
````

### Difference of Sets

````python
# even, but not in more_than_three set using -

print(even_numbers - more_than_three)
````

````python
# elements in more_than_three but not in even_numbers set using -

print(more_than_three - even_numbers)
````
