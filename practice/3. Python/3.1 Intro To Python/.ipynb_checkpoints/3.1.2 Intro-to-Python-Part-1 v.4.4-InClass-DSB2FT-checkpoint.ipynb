# Introduction to Data Types

In everyday life, numbers can have multiple interpretations. For example, "50 cm" is a measurement of length, while "50 kg" is a measurement of mass. Even though both values are 50, they mean very different things.

Similarly, computers understand data in certain ways, and this depends on the "type" of the data. 

The following basic types exist in Python:

* `int` - `0` (whole numbers)
* `float` - `1.25` (fractional numbers)
* `str` - `'hello'` (text)
* `bool` - `True` and `False` (conditions which have only two values)
* `NoneType` - `None` (the absence of meaningful information)

````python
# try to add 1 to '2', what would you get?

1+"ali"
````

From the above experiments, we can see that the type of data influences how we can act on it. For `ints` (numbers), the natural interpretation of `+` is addition. On the other hand, for `strs` (text), we instead perform *concatenation*, which simply means joining them together.

Now, when we try to use `+` on an `int` and a `str`, it won't work. This is because it doesn't make sense to add a number to text! If we want this to work, we have to change one of their types. This is termed *casting*:

### Variables restrictions and best practices:
**Restrictions**
- Variable names cannot be just a number (i.e., `2`, `0.01`, `10000`).
- Variables cannot be assigned the same name as a default or imported function (i.e., '`type`', '`print`', '`for`').
- Variable names cannot contain spaces.
- Variable name must start with a letter or the underscore character _
- Variable name cannot start with a number

**Best Practices**
- Variable names should be lowercase.
- A variable's name should be representative of the value(s) it has been assigned.
- If you must use multiple words in your variable name, use an underscore to separate them.

### Math exercise:
Recall the quadratic formula for solving a polynomial equation with coefficients $a$, $b$, $c$:

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$

````python
a = 1
b = -8
c = 15
````

````python
# write out the discriminant
discr = (b)**2 - (4*a*c)
````

````python
# Give me the code to produce one of the two roots

x = (((-(b))+(discr)**1/2)/2*(a))
````

````python
# display the string s

s
````

````python
# Slice objects at indexes 0, 1, 2, and 3

s[0:4]
````

Most ranges or functions with ranges have upper ends that are not inclusive. So, a range of `[0:5]` starts at `0` and stops before `5`.

````python
# show the s string from index 0 to index 5, What would you get?

s[0:5]
````

````python
# From index 6 up to the end of the string:

s[6::]
````

````python
# No start or end specified [:]:

s[0:]
````

````python
# Can we index from the right side? Yes!
# Try index -1

s[-1]
````

````python
# How about index -2? what does it return? guess!

s[-2]
````

````python
# How about index [-5:-1]? what does it return? 

s[-5:-1]
````

````python
# using find with slicing:
# find from string "this" until end

s[s.find("this"):]
````

````python
# using find with lower and upper slicing:
# find string "this" until string "FT"

s[s.find('this'):s.find("FT")]
````

<div>
  <h2 style="margin-top:0; color:#64b5f6; font-weight:700;">String Interpolation</h2>
</div>

<div style="background:#E3F2FD; border-left:6px solid #64B5F6; padding:12px 15px; border-radius:8px; margin:10px 0;"> <strong style="color:#1E88E5; font-weight:700;">Note:</strong> <span style="color:#0D47A1; display:block; margin-top:6px;">
  <span style="color:#0D47A1; display:block; margin-top:10px; font-size:15px;">
    <ul style="margin-top:4px; margin-left:20px; line-height:1.6;">
      <li>String interpolation allows you to substitute variable values inside strings.</li>
      <li>Done with <strong>f-strings</strong>, short for formatted strings.</li>
      <li>Prefix the string with the letter <code>f</code>.</li>
      <li>Surround the variable name with curly braces <code>{}</code> within the string.</li>
    </ul>
  </span>
</div>


````python
customer = "Mohammed"
month = "August"

# Make a message variable that equates 
# to "Hello Mohammed, this is your mobile bill of August." 
# using the variables customer and month
# using formatted string f"

message = "Hello Mohammed, this is your mobile bill of August."
print(f"Hello {customer}, this is your mobile bill of {month}.")
````

````python
# An alternative, more intuitive way of formatting a string using .format:

message = "Hello {} this is your mobile bill of {}".format(customer,month)

message
````

````python
# another example of using .format
# set the value of s3 to be a text "the pi is 3.1415" using .format 


s3= "the pi is {0}".format(3.1415)
s3
````

````python
# How about if we have multiple variables within a string:

s4 = "the pi is {0} while the radius is {1}".format(3.14,5.6)
s4
````

````python
# Create a variable product and save 'Technology' in it.
# Create a variable store and save 'Superstore' in it.
# Using the above defined variables, print the following string:
# Technology products in Superstore have a low return rate.

product = "Technology"
store = "Superstore"

print("{} products in {} have a low return rate.".format(product,store))
````
