# Introduction to Functions

Sometimes, you may have certain operations that you want to perform on multiple objects. For example, say you want to convert a number from Celsius to Fahrenheit, according to the following formula:

$$
F = \frac{9}{5} C + 32
$$

````python
# Do a code to convert the temperature 64 from C to F
first_temp_C = 64

````

````python
# Do a code to convert the temperature 36 from C to F
second_temp_C = 36

````

````python
# Do a code to convert the temperature 100 from C to F
third_temp_C = 100
````

That's a fair bit of repeated code. Imagine if we could, somewhere, write down the method to transform C into F once, and just apply it the same way all three times! That's exactly what a function does:

This code you could probably simplify using a function. 

In Python, functions can be constructed along the following rules.


* Function blocks begin with a single line including in that order
    * the keyword **`def`**  
    * the function name 
    * parentheses ( `( )` ) which may or may not include parameters; parameters can be assigned inside these parentheses, too
    * and a colon (`:`)
* The following lines of the function block include what the function is supposed to do. All these lines have to be indented.
* The first statement of a function can be an optional statement - the documentation string of the function or **docstring**. It can be called with `?functionname` or `help(functionname)`.
* The statement `return [expression]` exits a function, optionally passing back an expression to the caller. A return statement with no arguments is the same as `return None`.

The general syntax for a function is indicated below:

```python
def functionname( parameters ):
    "function_docstring"
    function_suite
    return expression
```

### Docstring

Lastly, a function can have something called a *docstring*, which acts as documentation, or an explanation of what a function does. Let's define the Celsius to Fahrenheit function again, this time with a docstring:

````python
def doc_convert(celsius):
    "Convert Celsius to Fahrenheit according to the equation F = 9 / 5 * C + 32."
    return 9 / 5 * C + 32
````

You can access the docstring with `help` or `?`:

````python
# using help() to get the function's documentation
help(doc_convert)
````

````python
# using "?" to get the function's documentation 
# use function?
doc_convert?
````

````python
# now try the most known function `print`
# print?
# help(print)
````

When providing arguments to a function, you can either rely on the position or on keywords to define the specific inputs.


````python
def calculate_area(length, width):
    return length * width
````
