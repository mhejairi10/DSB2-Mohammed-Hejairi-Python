## Introduction: if/elif/else statements

An if statement is a fundamental control structure in programming languages that allows you to execute specific code blocks based on whether a condition is true or false. It is used to make decisions and control the flow of execution in your program.

<a name="demo_if"></a>
## Python `if` Statement
The general Python syntax for a simple if statement is:
```python
        if condition:
            indentedStatementBlock
```

If the condition is true, then execute the indented statements. If the condition
is not true, then skip the indented statements.

>*Note: Be mindful of the indentation while writing the if statements. Indentation is the whitespace at the beginning of the code.*



An example is given below. Try values which lead to different results in the if-statement.

## Python `if...else` Statement

Sometimes we want a certain statement to be executed when a given condition is met, and another statement to be executed if the condition is not met. That can be obtained with **if-else** statements.

The general Python if-else syntax is

```python
        if condition:
            indentedStatementBlockForTrueCondition
        else:
            indentedStatementBlockForFalseCondition
```

These statement blocks can have any number of statements, and can include
about any kind of statement.
:

### Examples: Python `if...else` Statement

#### *Example1*:

````python
# Note: the indentation is a must in if statement
number = 10
if number > 0:
    print("positive")
else:
    print("Negative")

````

>*Note that now there are two indented blocks. If the `if-condition` is met, the else part is never executed. If the `if-condition` is not met, the else condition will automatically be executed.*

````python
# Do an else statement for the previous suitcase weight example

weight = input("enter suitcase weight: ")
if eval(weight) > 23:
    print("extra $25 charge")
else:
    print("no extra charge")

````

## Python `if…elif…else` Statement

Sometimes we want to distinguish more than two cases. This can be done with **if/elif/else** statements. 
The syntax for an if-elif-else statement is indicated in general below:

```python
        if condition1:
            indentedStatementBlockForFirstTrueCondition1
        elif condition2:
            indentedStatementBlockForFirstTrueCondition2
        elif condition3:
            indentedStatementBlockForFirstTrueCondition3
        elif condition4:
            indentedStatementBlockForFirstTrueCondition4
        else:
            indentedStatementBlockForEachConditionFalse
```

The first condition is checked with an if-statement, all following conditions are checked by an elif statement, if none of the previous conditions has been met, the else statement is executed. 

Note that the conditions are checked in the given order. Once a condition is met, the later ones will not be verified. For this reason it is important that the different conditions do not overlap. If you think condition 3 is met, but you notice that the block from condition 2 is executed, the reason is probably that both conditions 2 and 3 are met. 

Elif and else statements are optional, but the first statement has to be an if-statement.
The if, each elif, and the final else line are all aligned. There can be any number
of elif lines, each followed by an indented block. If all three are present, exactly
one of the indented blocks is executed. It is the one corresponding to the first
True condition.

Let's do an example:

### Examples: Python `if...elif...else` Statement

````python
# Do if elif else with the example of suitcase weight example

weight = input("enter suitcase weight: ")
if eval(weight) > 23:
    print("extra $25 charge")
elif eval(weight) <=0:
   print("no value")
else:
    print("no extra charge")
    
````

## Note on using parentheses `()` in conditional statements
- Some programming languages require the `condition` to be enclosed in parentheses, but Python does not
- However, they can enhance readability and clarity in certain situations, like in the following codes, which is more clear?

  ```python

  if a > b and b < c or d == e:
    print("Condition met")

      
  if (a > b and b < c) or (d == e):
    print("Condition met")
  ```
  ```

- Also, logical operators have precedence rules, which can lead to **unexpected results** if parentheses are not used. Explicitly grouping expressions with parentheses helps reinforce the intended order of evaluation. Check the following code:
  ```python
    x = True
    y = False
    z = True

    print(x or y and not z) # Prints True

    print((x or y) and not z) # Prints False
  ```
  ```
- In Python, the order of the higher precedence is `not` then `and` then `or`

## Python Nested Conditionals 
A nested if statement is one that has an if statement within another if statement. 

```python
if condition1:
    if condition2:
        # code block to execute if both condition1 and condition2 are True
    else:
        # code block to execute if condition1 is True but condition2 is False
else:
    # code block to execute if condition1 is False
```
``` False

````python
# doing a nested if statement

product_is_good = True
bank_account = 500

if product_is_good == True:
    if bank_account > 100:
        print("let's buy this product")



````

### Multiple Conditions, One Statement

- You can check to see if two conditions are met in one statement with a logic operator



````python
# Can we do the previous nested if statement in one conditional statement?

product_is_good = True
bank_account = 500

if product_is_good == True and bank_account > 100:
        print("let's buy this product")

````

## Condensing Conditionals

not all conditionals need a comparison statement — especially if the values being tested are already Booleans.
Thus, you will rarely see a comparison with “== true” or “!== false”.
Instead, you can use the following pattern:

````python
# Example on Condensing Conditionals

product_is_good = True
product_not_expensive = True

if product_is_good and product_not_expensive:
        print("let's buy this product")


````
