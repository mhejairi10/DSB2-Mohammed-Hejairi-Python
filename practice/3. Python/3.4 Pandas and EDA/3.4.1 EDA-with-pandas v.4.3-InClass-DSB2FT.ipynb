### DataFrame vs. Series

---
Pandas relies on two key objects, each with their own methods and properties:

* A **`Series`** is a one-dimensional array of values (**`one column`**) that contains a row index.

* A **`DataFrame`** is a two-dimensional array of values **(`table`)** *with both a row and column index*.
    * It turns out - each column of a `DataFrame` is actually a `Series`!



There is an important difference between using a list of strings versus only using a string with a column's name: When you use a list containing the string, it returns another **DataFrame**. But, when you only use the string, it returns a `pandas` **Series** object.

## Vectorized operations
---

Why bother with `DataFrame` and `Series` object when we could use Python lists and dictionaries?

`pandas` uses `numpy` arrays under the hood, which can perform operations **on an entire collection of values** rather than looping through the values individually.

Let's see a basic example.

We'll square the first 100,000 integers: first with a `for` loop, then as a `pandas` `Series` and compare timings.

*(we can use Jupyter's `%%timeit` command to time how long a cell takes to run)*

````python
numbers = range(100000)
numbers
````

Option 1: Using `for` loop

````python
%%timeit

squared_numbers = []

for number in numbers:
    squared_numbers.append(number**2)
````

Option 2: Using `pandas` `Series`

````python
%%timeit

number_series = pd.Series(numbers)
squared_numbers = number_series**2
````

# Exploratory Data Analysis (EDA)
---

In a nutshell, **Exploratory Data Analysis (EDA)** means “getting to know” a data set.

This can include:
- Checking data types to make sure data is stored properly.
- Calculating summaries for columns, like the average, minimum, or maximum.
- Evaluating your data set for missing data.
- Identifying potential trends or outliers.
- Basic visualization of your data.


It's common to reach a later point in the Data Analytics Workflow only to realize that unclean data or a particular feature could have be engineered earlier in the process. Exploring the data first makes working with it later much more reliable. Hypothesis-driven EDA is essential for effective EDA, otherwise, we would be endlessly mining our data for answers.

### Exploratory Data Analysis: Best Practices

At the very least, as part of EDA, you should determine:
- The number of rows in the data set.
    - What does each row represent? Is each row a person, an observation, a time point?
- The number of columns in the data set.
    - What does each column represent? How was that data collected? Try using a data dictionary — it can often directly answer these questions for you!

One of the most challenging parts of data analytics can be turning business questions into analyses, or even forming your own questions when the request you receive is vague. You can start by asking these two questions:

- What fields can I COMBINE to find interesting insights?
- What ACTIONS can someone take as a result of my charts and analyses?

Answering them will also help you figure out what to do with your projects.

## 1. Examining Your Data Types With `.info()`

---

When getting acquainted with a new data set, `.info()` should be **the first thing** you examine.

**Types** are very important. They affect the way data will be **represented** in our machine learning/data analyses, how data can be joined, whether or not math operators can be applied, and instances in which you can encounter unexpected results.

> _Typical problems that arise when working with new data sets include_:
> - Missing values.
> - Unexpected types (string/object instead of int/float).
> - Dirty data (commas, dollar signs, unexpected characters, etc.).
> - Blank values that are actually "non-null" or single white-space characters.

`.info()` is a function available on every **DataFrame** object. It provides information about:

- The name of the column/variable attribute.
- The type of index (RangeIndex is default).
- The count of non-null values by column/attribute.
- The type of data contained in the column/attribute.
- The unique counts of **dtypes** (`pandas` data types).
- The memory usage of our data set.

````python
# Show the info() for the DataFrame

orders.info()
````

- Note: Strings are stored as a type called "object," as they are not guaranteed to take up a set amount of space (strings can be any length).

- You can also check the columns data types using `dtypes`

## 4. Dataframe Filtering

Filtering and sorting are important steps that allow us to drill into subsets of our data. To filter, we use a process called Boolean filtering, wherein we first define a Boolean mask and then use it to filter our DataFrame.

Filtering is how we start to split our data and generate meaningful insights based on the differences in various subsets.
While basic exploratory methods are nice, more insights come from splitting, combining, and comparing subsections of data.


````python
orders.head()
````

### 4.1 Filtering by Single Condition

````python
# filtering one column and showing the filter result
# show the rows that has 'Same Day' in the 'ship_mode' column
orders['ship_mode'] == 'Same Day'
````

````python
# using Boolean mask to filter the data we want and show all the dataframe

orders[orders['ship_mode'] == 'Same Day'].head()
````

````python
# show only single column 'profit' for the filtered data

orders[orders['ship_mode'] == 'Same Day'][['profit']]
````

Once we have a Boolean Series, we can then filter the entire data set for only those values with a True result. Apply this mask to the DataFrame using the syntax: `dataframe[boolean mask]`

When filtering with this syntax, the result is a DataFrame. This means that we can continue accessing other columns or using methods.

Chain on the column name we're interested in, `['profit']`, and the `.mean()` function to get the mean profit for orders that have `Same Day` ship mode.

````python
# This will get the mean for all 'profit' column

orders['profit'].mean()

````

````python
# This will get the mean of the 'profit' column for the filtered data
orders[orders['ship_mode'] == 'Same Day'][['profit']].mean()

````

````python
# find all the rows that have customer id "BT-11305"

orders[orders["customer_id"] == "BT-11305"]
````

````python
# find all the orders that have quantity > 13
orders[orders["quantity"]>13]
````

````python
# find all the rows that is higher than the average profit

orders[orders["profit"] > orders['profit'].mean()]

````

````python
# find all the rows that has the maximum profit
# Note: orders['profit'].max() will get the maximum profit
orders[orders['profit'] == orders['profit'].max()]
````

### 4.2 Filtering by Multiple Conditions

Often,we want to filter based on multiple conditions. 
- In pandas, we have to use the symbol:
    - `&` to make "and" in a logical condition
    - `|` to make "or" in a logical condition
- Note that you also **`must use parentheses`** to separate your conditions.

````python
# find all the orders that have quantity > 13 for customer "AG-10495"

orders[(orders['quantity'] > 13) & (orders['customer_id'] == "AG-10495")]

````

````python
# find all the orders for either customers "AG-10495" or customer JB-16045"
# that have sales > 500 

orders[((orders['customer_id'] == "AG-10495") 
        | (orders['customer_id'] == "JB-16045")) &
        (orders['sales'] > 500)]
````

### 4.3 Filtering Using Part of a String


Often,we want to filter based on specific string in column(s). 
- The following are used for string filteration:
    - `.str.contains()  `: Checks if each element in the Series/Column contains a specified substring.
    - `.str.startswith()`: Checks if each element starts with a specified substring.
    - `.str.endswith()  `: Checks if each element ends with a specified substring
> Note: all of the the above string search methods are case sensitive. To be able to search for any text regardless of case sensitivity, make everything in lower case then search. Example:   `.str.lower().str.contains() `

````python
# show the orders that have product_id that contains "FU-10"
# show only 5 header result

orders[orders['product_id'].str.contains("FU-10")]
````

````python
# show the orders that have customer_id starts with "TT"
# show only 5 header result

orders[orders['customer_id'].str.startswith("TT")]
````

````python
# show the orders that have product_id ends with with "508"
# show only 5 header result

orders[orders['product_id'].str.endswith("508")]
````

### 4.4 Filtering Using Part of a String with multiple conditions

Again:
- In pandas, we have to use the symbol:
    - `&` to make "and" in a logical condition
    - `|` to make "or" in a logical condition
- Note that you also **`must use parentheses`** to separate your conditions.

````python
# show the orders that have product_id that contains "FU-10" 
# and customer_id contains "CS-12"
# show only 5 header result

orders[(orders['product_id'].str.contains("FU-10")) & 
    (orders['customer_id'].str.contains("CS-12"))].head()
````
