## Common metrics for evaluating regression models

---

### 1. Sum of Squared Errors (SSE)

**Sum of Squared Errors (SSE)**, also known as the Residual Sum of Squares (RSS), is a metric used to measure the total deviation of data points from a regression model's predicted values. It quantifies the amount of variance in a dataset that the model fails to explain.

#### Mathematical Formula

The formula for calculating SSE is:

$$SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:
*   $y_i$ is the actual observed value.
*   $\hat{y}_i$ is the predicted value generated by the model.
*   $n$ is the total number of data points.

#### Key Characteristics

*   **Error Measurement**: It squares each residual (the difference between actual and predicted values) to eliminate negative signs so that overestimations and underestimations do not cancel each other out.
*   **Sensitivity to Outliers**: Because it squares the errors, mistakes contribute disproportionately more to the final score than smaller ones, making SSE sensitive to extreme outliers.
*   **Optimization Goal**: In Ordinary Least Squares (OLS) linear regression, the primary goal of the algorithm is to minimize this specific value.
*   **Differentiability**: SSE is highly popular in machine learning optimization (like gradient descent) because its mathematical derivative is easy to compute.

#### Practical Interpretation

*   **SSE = 0**: The model perfectly predicts every single data point in the dataset.
*   **Lower SSE**: Indicates a tighter, more accurate fit of the model to the training data.
*   **Higher SSE**: Indicates poor predictive performance and significant distance between the model's line and the actual data points.



### 2. Mean Squared Error (MSE)

**Mean Squared Error (MSE)** is a standard metric used to measure the average performance of a regression model. While SSE calculates the total accumulated error, MSE divides that total by the number of data points to find the average squared distance between predicted and actual values.

#### Mathematical Formula

The formula for calculating MSE is:

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Alternatively, you can express it in relation to SSE:

$$MSE = \frac{SSE}{n}$$

*   **Average Penalty**: It measures variance on a per-sample basis, making it easier to compare model performance across datasets of different sizes.



### 3. Root Mean Squared Error (RMSE)

**Root Mean Squared Error (RMSE)** is an extension of the Mean Squared Error (MSE) that measures the square root of the mean of the sum of squared `residuals`. Unlike MSE, RMSE is expressed in the exact same units as the target variable (e.g., if predicting house prices in dollars, RMSE is directly interpreted in dollars).

#### Mathematical Formula

The formula for calculating RMSE is:

$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

Alternatively, you can express it directly using MSE:

$$RMSE = \sqrt{MSE}$$

The smaller the root mean squared error, the better your model fits the data. 

**Calculate the RMSE of your model by leveraging `sklearn.metrics.mean_squared_error`.**

### 4. R-squared ($R^2$ or Coefficient of Determination)

**$R^2$ is the most common metric to evaluate a regression and is the default scoring measure in sklearn.**
While SSE, MSE, and RMSE measure error magnitude, $R^2$ evaluates the explanatory power of the model.

#### Mathematical Formula

The formula for calculating R-squared is:

$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$

Where:
*   $SS_{res}$ or $SSE$ is the **Sum of Squared Errors** (residual sum of squares) for our model:

    $SS_{res}=\sum_i (y_i - \hat{y})^2$
*   $SS_{tot}$ is the **Total Sum of Squares**, which measures the total variance in the actual data relative to the mean ($\bar{y}$):

    $SS_{tot} = \sum_i (y_i-\bar{y})^2$
*   $y_i$ is the actual value, $\hat{y}_i$ is the predicted value, and $\bar{y}$ is the mean of the actual values.

#### Key Characteristics

*   **Standardized Scale**: Unlike MSE or RMSE, $R^2$ is generally scale-independent and ranges from $0$ to $1$ (though it can be negative if the model is worse than a simple horizontal line at the mean).
*   **Baseline Comparison**: It compares your model's performance against a baseline model that always predicts the mean value of the dataset.
*   **Overfitting Risk**: $R^2$ will always increase or stay the same when new predictors are added to the model, even if those predictors add no real value. This limitation is typically corrected using **Adjusted R-squared**.

#### Practical Interpretation

*   **$R^2 = 1$**: The model explains 100% of the variance in the target variable (perfect prediction).
*   **$R^2 = 0$**: The model performs no better than predicting the mean value for every data point.
*   **$R^2 = 0.75$**: Approximately 75% of the observed variation in the data can be explained by the model's inputs.


### 5. Regression Metrics Comparison Table

A comprehensive summary table comparing the evaluation metrics: 

| Metric | Full Name | Mathematical Formula | Primary Unit | Sensitivity to Outliers | Best Ideal Value |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **SSE** | Sum of Squared Errors | $$\sum (y_i - \hat{y}_i)^2$$ | Squared units of target variable | **Extremely High** (Increases with data size) | `0` (Perfect Fit) |
| **MSE** | Mean Squared Error | $$\frac{1}{n} \sum (y_i - \hat{y}_i)^2$$ | Squared units of target variable | **High** (Penalizes larger errors heavily) | `0` (Perfect Fit) |
| **RMSE** | Root Mean Squared Error | $$\sqrt{\frac{1}{n} \sum (y_i - \hat{y}_i)^2}$$ | Same units as target variable | **High** (Retains penalty structure of MSE) | `0` (Perfect Fit) |
| **$R^2$** | Coefficient of Determination | $$1 - \frac{SS_{res}}{SS_{tot}}$$ | Dimensionless (Scale-independent ratio) | **Moderate** (Influenced via underlying SSE) | `1` (Perfect Explanation) |

#### Critical Insights for Selection

*   Use **RMSE** when you need to communicate average prediction error directly to non-technical stakeholders in native units.
*   Use **MSE** during model optimization or loss function calculations because it handles mathematical derivatives smoothly.
*   Use **$R^2$** when comparing the fit of models across completely different datasets, as it normalizes performance onto a standardized scale.

```python
# Create the design matrix (X) and response vector (y)
X = ads.drop(columns="sales")
y= ads["sales"]
```

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(X,y)
```

```python
# Create a prediction vector y_hat

y_hat = model.predict(X)
```

```python
print(model.coef_,model.intercept_)
```

# Metric #2: Sum Squared Error (SSE)


```python
# Compute the SSE using numpy
sse = np.sum((y-y_hat)**2)
print(sse)
```

# Metric #3: Mean Squared Error (MSE)


```python
# Compute the MSE using numpy
mse = np.mean((y-y_hat)**2)
print(mse)
```

```python
# Compute the MSE using sklearn!
metrics.mean_squared_error(y,y_hat)
```

### Metric #3.5: The Root Mean Squared Error (RMSE)


```python
# Compute the RMSE! Does it make sense?
np.sqrt(metrics.mean_squared_error(y,y_hat))
```

RMSE is the square root of MSE, bringing the error metric back to the original target units, making it more interpretable than MSE, which uses squared units
