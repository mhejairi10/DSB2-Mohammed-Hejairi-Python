# Connecting to APIs

### What is an API?

**A**pplication **P**rogramming **I**nterface:

API is how computers communicate. It's a connection point, typically over the internet, that allows two computers to interact. For our purposes, it provides a safe and legal way to access large amounts of structured data from the web, whether owned by private companies, governments, or public organizations. 

APIs are used to send and receive data, and they define what data can be exchanged and the commands/syntax used.

### Why use APIs?

Many organizations rely on the web as a key source of data:
- Agricultural commodity traders want real-time access to weather data via web APIs as weather conditions affect the supply of the commodity, and therefore the price.
- Reliability engineers at mine sites want to access the data streamed off ore processing machinery then stored in “the cloud”
- Market research firms use the Twitter APIs to uncover important consumer trends.

### How does it work?

![from g2.com](https://learn.g2.com/hs-fs/hubfs/G2CM_FI167_Learn_Article_Images_%5BAPI%5D_Infographic_V1a.png)

#### Steps:
1.	Request:
    -	A client (You) sends a request to an API server through the internet. 
    -	This request is in a form of a URL with specific parameters, indicating the desired action. In some instances, it may include authentication credentials to verify the client's identity.
2.	Process:
    -	The API server receives the request and processes it.  It interprets the request, determines the appropriate action, and fetches the necessary data from its database or other sources.
3.	Response:
    -	The API server sends a response back to the client. The response typically contains the requested data in a specific format, such as JSON or XML.  It may also include additional information like error codes or success messages.


**👉** We do need to be aware of __[HTTP status codes](https://en.wikipedia.org/wiki/List_of_HTTP_status_codes)__ provided by the API that will inform us if our requests:

|Code|Reason|
|---|-----|
|200|Succeed|
|400 - 599|Failed|
|404|Does not exist|

````python
omdb_data = omdb_req.json()                                                 #--> Gettoing the data in JSON format from the request's response
pprint(omdb_data)
````

````python
# Convert the data to DataFrame
df_movie = pd.DataFrame([omdb_data])
df_movie.head()
````

````python
# Let's explore the first record/observation
df_movie.iloc[0]
````

__Notice__ the 'Ratings' column, it contains a list. This is something common you would face when dealing with some datasets, specially JSON formatted.

````python
# Looking at the 'Ratings' value
df_movie.loc[0,'Ratings']
````

__Note :__ The columns Internet Movie Database, Rotten Tomatoes, and Metacritic and inside the Ratings column

Instead of having them all in a column, we can add each in a new column

````python
# Create a new column 'Internet Movie Database' and add the rating
df_movie['Internet Movie Database'] = df_movie['Ratings'].iloc[0][0]['Value']
df_movie['Internet Movie Database'] = df_movie['Internet Movie Database'].str.split('/').str[0]
df_movie['Internet Movie Database'] = df_movie['Internet Movie Database'].astype(float)
df_movie['Internet Movie Database']
````

````python
# Create a new column 'Rotten Tomatoes' and add the rating
df_movie['Rotten Tomatoes'] = df_movie['Ratings'].iloc[0][1]['Value'] 
df_movie['Rotten Tomatoes']  = df_movie['Rotten Tomatoes'].str[0:-1]
df_movie['Rotten Tomatoes'] = df_movie['Rotten Tomatoes'].astype(float)
df_movie['Rotten Tomatoes']
````
